```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy.io import wavfile
import wave
```

```python
file = wave.open("output.wav", "r")
file1 = wave.open('data/mini_speech_commands/no/0ab3b47d_nohash_0.wav')

signal = file.readframes(-1)
signal = np.fromstring(signal, dtype=np.int16)

signal1 = file1.readframes(-1)
signal1 = np.fromstring(signal1, dtype=np.int16)

plt.figure(1)
plt.title("Waveform")
plt.plot(signal)
plt.show()

plt.figure(1)
plt.title("Working Waveform")
plt.plot(signal1)
plt.show()
```
